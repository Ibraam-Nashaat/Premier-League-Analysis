```python
# import libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import random 
```

```python
# loading file
df=pd.read_csv("eplmatches.csv")
# show the data
df
```

```python
# GET POINTS FOR EACH TEAM PLAYED IN MATCH
df['HT_points'] = np.where(df['FTR'] == 'H',3, np.where(df['FTR'] == 'A', 0, 1))
df['AT_points'] = np.where(df['FTR'] == 'H',0, np.where(df['FTR'] == 'A', 3, 1))
df.head()
```

```python
home_team_win_df = df.groupby("Home").agg({
        "HT_points": "sum",
    })

home_team_win_df.sort_values(by= 'HT_points',ascending=False)
```

```python
away_team_win_df = df.groupby("Away").agg({
        "AT_points": "sum",
    })

away_team_win_df.sort_values(by= 'AT_points',ascending=False)
```

```python
home_points_mean=home_team_win_df['HT_points'].mean()
away_points_mean=away_team_win_df['AT_points'].mean()
diff_points_mean=home_points_mean-away_points_mean
diff_points_mean
```

```python
home_team_win_df.count()
```

```python
home_team_wins_renamed=home_team_win_df.copy()
home_team_wins_renamed.columns = ['HT_points']
home_team_wins_renamed=home_team_wins_renamed.reset_index()
home_team_wins_renamed.rename(columns = {'Home':'Team','HT_points':'points'},inplace=True)
home_team_wins_renamed
```

```python
away_team_wins_renamed=away_team_win_df.copy()
away_team_wins_renamed.columns = ['AT_points']
away_team_wins_renamed=away_team_wins_renamed.reset_index()
away_team_wins_renamed.rename(columns = {'Away':'Team','AT_points':'points'},inplace=True)
away_team_wins_renamed
```

```python
team_wins=pd.concat([home_team_wins_renamed,away_team_wins_renamed], ignore_index=True)
team_wins
```

```python
def perm_fun(x, nA, nB):
    n = nA + nB
    idx_B = set(random.sample(range(n), nB))
    idx_A = set(range(n)) - idx_B
    return x.loc[list(idx_B)].mean() - x.loc[list(idx_A)].mean()
```

```python
perm_diffs = [perm_fun(team_wins['points'], 50, 50) for _ in range(2000)]
diff= np.mean(np.array(perm_diffs) > diff_points_mean)
print("{0:.3f}".format(diff))
```
